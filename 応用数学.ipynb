### レポート本文
本章の目標は、1) 条件付き確率について理解を深める。2) ベイズ則の概要を知る。3) 期待値・分散の求め方を確認する。4) 様々な確率分布の概要を知る。であった。以下、目標ごとに要点を整理する。
1. 条件付き確率について理解を深める。
条件付き確率はある事象X=xが与えられたもとでY=yとなる確率のこと。例えば、雨が降っている条件下で交通事故にあう確率は、<br>
(雨が降っている かつ 交通事故に合う確率) / (雨が降っている確率)で求められる。 
2. ベイズ則の概要を知る。
頻度確率が発生する頻度を意味するのに対し、ベイズ確率は信念の度合いを意味する。主観確率とも呼ばれる。ベイズ則は、事象X=xと事象Y=yに対して<br>
P(X=x|Y=y)P(Y=y) = P(Y=y|X=x)P(X=x)となる。<br>
ベイズ統計では事前確率を設定し、観測された追加の情報をもとに事前確率を更新していく。ここで設定する事前確率は主観に基づく確率でもよいとされている。頻度確率と比較したメリットの一つに、データ量が少ない場合でも適用可能な点が挙げられる。（『完全独習　ベイズ統計学入門』参照）
